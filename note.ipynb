```python
import numpy as np
import pandas as pd
```

```python
mv = pd.read_csv('dataset/tmdb_5000_movies.csv')
cr = pd.read_csv('dataset/tmdb_5000_credits.csv')
```

```python
mv.head()
```

```python
cr.head()
```

```python
mv = mv.merge(cr,on='title')
```

```python
mv.head(1)
```

```python
mv_original = mv.copy()
```

```python
mv = mv[['id', 'genres', 'keywords', 'title', 'overview', 'cast', 'crew']]
```

```python
mv.head(1)
```

```python
import ast
```

```python
def convert(obj):
    l = []
    for i in eval(obj):
        l.append(i['name'])
    return l
```

```python
mv['genres'] = mv['genres'].apply(convert)
```

```python
mv['keywords'] = mv['keywords'].apply(convert)
```

```python
def convert_cast(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
mv['cast'] = mv['cast'].apply(convert_cast)
```

```python
mv.head(1)
```

```python
mv = mv[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
mv.head(1)
```

```python
mv['crew'][0]
```

```python
mv['crew'].unique()
```

```python
def convert_dir(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] =='Director':
            l.append(i['name'])
            break
    return l
```

```python
mv['crew'] = mv['crew'].apply(convert_dir)
```

```python
mv.head(1)
```

```python
def remove_space(l):
    new_l=[]
    for i in l:
        new_l.append(i.replace(" ", ""))
    return new_l
```

```python
mv['genres'] = mv['genres'].apply(remove_space)
mv['keywords'] = mv['keywords'].apply(remove_space)
mv['cast'] = mv['cast'].apply(remove_space)
mv['crew'] = mv['crew'].apply(remove_space)
```

```python
mv.head(1)
```

```python
mv['overview'] = mv['overview'].apply(lambda x:x.split())
```

```python
mv['overview'].isna().sum()
```

```python
mv['overview'].dropna(inplace=True)
```

```python
mv['overview'].isna().sum()
```

```python
mv['overview'] = mv['overview'].fillna('').apply(lambda x:x.split())
```

```python
mv.head(1)
```

```python
mv['tags'] = mv['overview'] + mv['genres'] + mv['keywords'] + mv['cast'] + mv['crew']
```

```python
mv.head(1)
```

```python
new_mv = mv[['id', 'title', 'tags']]
```

```python
new_mv.head(1)
```

```python
new_mv['tags'][0]
```

```python
new_mv['tags'] = new_mv['tags'].apply(lambda x: " ".join(x))
```

```python
new_mv['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_mv['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity  = cosine_similarity(vectors)
```

```python
new_mv.head(1)
```

```python
similarity[0]
```

```python
similarity[3]
```

```python
new_mv[new_mv['title'] == 'Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])
```

```python
def recommend(movie):
    idx = new_mv[new_mv['title'] == movie].index[0]
    distance = similarity[idx]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_mv.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
